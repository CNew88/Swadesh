```python
import pandas as pd
import numpy as np
import os
```

### Goal 1: Calculate the average edit distance between each pair of languages

```python
## Path to directory where .csv files are stored locally
csv_path = "/Users/brookehamilton/Dropbox/Rochester/linguistics/swadesh/asjp_dataset.cldf/"
```

```python
## Read all filenames from the directory of csvs
## Append each to a list

path_list = []
for filename in os.listdir(csv_path):
    if filename.endswith(".csv"): 
        path_list.append((os.path.join(csv_path, filename)))  
```

```python
path_list
```

```python
list_of_dfs = []  # A list to store each dataframe object in

for i in path_list:
    this_df = pd.read_csv(i)
    list_of_dfs.append(this_df)
    
```

```python
forty_words = ["I", "you", "we", "one", "two", "person", "fish", "dog", "louse", "tree", 
               "leaf", "skin", "blood", "bone", "horn", "ear", "eye", "nose", "tooth",
              "tongue", "knee", "hand", "breast", "liver", "drink", "see", "hear", "die",
              "come", "sun", "star", "water", "stone", "fire", "path", "mountain", "night",
              "full", "new", "name"]
```

```python
bafia = list_of_dfs[0]  # Here is what the first dataframe looks like
```

```python
bafia
```

```python
## To flatten the data frame, we have to drop duplicates
bafia = bafia.drop_duplicates(['Language_name','Parameter_name'])
```

```python
## Flatten using "pivot"
bafia.pivot(index="Language_name", columns='Parameter_name', values='Value')
```

```python
## For each language, drop duplicates, then flatten
for i in range(len(list_of_dfs)):
    list_of_dfs[i] = list_of_dfs[i].drop_duplicates(['Language_name','Parameter_name'])
    list_of_dfs[i] = list_of_dfs[i].pivot(index="Language_name", 
                                          columns='Parameter_name', values='Value')  
```

```python
## Confirm that it worked
list_of_dfs[1]
```

```python
## Concatenate each 1-row df to a new data frame with all languages
full_df = pd.concat(list_of_dfs)
```

```python
## Save to file
full_df.to_csv("full_df.csv")
```
